Time series plots of interannual variability of surface temperature, salinity, DIN, Chl-a, upper mixed layer depth using monthly output for the period 1995-2005.

To download the data, we should split it in two parts of ten years each,
to avoid problems with the header notations from the last years

```python
#sst
m_list = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
sst1=np.zeros((120,180,216))
va=0
for i in range(1995,2013): 
    for m in m_list:
        fyd=nc4.Dataset('/group_workspaces/jasmin2/nemo/vol1/ORCA0083-N006/means/'+str(i)+'/ORCA0083-N06_'+str(i)+'m'+str(m)+'T.nc','r',format='NETCDF4')
        sst_=fyd.variables['sst'][0,1373:1493,3865:4045]
        sst1[:,:,va]=sst_
        va=va+1

sst2=np.zeros((120,180,36))
va=0
for i in range(2013,2016): 
    for m in m_list:
        fyd=nc4.Dataset('/group_workspaces/jasmin2/nemo/vol1/ORCA0083-N006/means/'+str(i)+'/ORCA0083-N06_'+str(i)+'m'+str(m)+'T.nc','r',format='NETCDF4')
        sst2_=fyd.variables['tos'][0,1373:1493,3865:4045]
        sst2[:,:,va]=sst2_
        va=va+1        
              
        
sst1[np.abs(sst1) > 3000.] = np.nan 
sst1[sst1 == 0.] = np.nan
sst2[np.abs(sst2) > 3000.] = np.nan 
sst2[sst2 == 0.] = np.nan

sst=np.concatenate((sst1,sst2),axis=2)
```

```python
#Salinity and Surface Layer Depth
m_list = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
sss1=np.zeros((120,180,216))
va=0
for i in range(1995,2013): 
    for m in m_list:
        fyd=nc4.Dataset('/group_workspaces/jasmin2/nemo/vol1/ORCA0083-N006/means/'+str(i)+'/ORCA0083-N06_'+str(i)+'m'+str(m)+'T.nc','r',format='NETCDF4')
        sss_=fyd.variables['sss'][0,1373:1493,3865:4045]
        sss1[:,:,va]=sss_
        va=va+1

sss2=np.zeros((120,180,36))
va=0
for i in range(2013,2016): 
    for m in m_list:
        fyd=nc4.Dataset('/group_workspaces/jasmin2/nemo/vol1/ORCA0083-N006/means/'+str(i)+'/ORCA0083-N06_'+str(i)+'m'+str(m)+'T.nc','r',format='NETCDF4')
        sss2_=fyd.variables['sos'][0,1373:1493,3865:4045]
        sss2[:,:,va]=sss2_
        va=va+1        
              
        
sss1[np.abs(sss1) > 3000.] = np.nan 
sss1[sss1 == 0.] = np.nan
sss2[np.abs(sss2) > 3000.] = np.nan 
sss2[sss2 == 0.] = np.nan

sss=np.concatenate((sss1,sss2),axis=2)

####
mld=np.zeros((120,180,252))
va=0
for i in range(1995,2016): 
    for m in m_list:
        fyd=nc4.Dataset('/group_workspaces/jasmin2/nemo/vol1/ORCA0083-N006/means/'+str(i)+'/ORCA0083-N06_'+str(i)+'m'+str(m)+'T.nc','r',format='NETCDF4')
        mld_=fyd.variables['mldkz5'][0,1373:1493,3865:4045]
        mld[:,:,va]=mld_
        va=va+1

mld[np.abs(mld) > 3000.] = np.nan 
mld[mld == 0.] = np.nan
```
